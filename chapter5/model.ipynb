```python
'''使用GPU进行推理'''
from transformers import AutoTokenizer, AutoModel
import torch
tokenizer = AutoTokenizer.from_pretrained("../models/bert-base-cased")
model = AutoModel.from_pretrained("../models/bert-base-cased", device_map="cuda")

raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]
sequence = "Using a Transformer model is simple"

inputs = tokenizer(sequence, padding=True, truncation=True, return_tensors="pt")
inputs = {k: v.to('cuda') for k, v in inputs.items()} 

print("输入的 input_ids 形状:", inputs["input_ids"].shape)
# print(inputs)
with torch.no_grad():
    outputs = model(**inputs)

print("输出类型:", type(outputs))
last_hidden = outputs.last_hidden_state
print("last_hidden_state 形状:", last_hidden.shape) 
pooler = outputs.pooler_output
print("pooler_output 形状:", pooler.shape) 

# 5. 如果你想要每层的隐藏状态或注意力权重，需要在调用时传入参数
outputs_full = model(
    **inputs,
    output_hidden_states=True,    # 获取所有层的隐藏状态
    output_attentions=True        # 获取所有层的注意力权重
)

# 所有层的隐藏状态（包括 embedding 层）
hidden_states = outputs_full.hidden_states
print("隐藏状态层数:", len(hidden_states))       # 13 (embedding + 12 layers)
print("第 0 层 (embedding) 形状:", hidden_states[0].shape)  # (batch, seq_len, 768)
```
