```python
from torch import nn

class Module1(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.lay1 = nn.Linear(3,5)
        self.activation = nn.ReLU()
        self.lay2 = nn.Linear(5,3)
    
    def forward(self, x):
        x = self.lay1(x)
        x = self.activation(x)
        x = self.lay2(x)
        return x

class Module2(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.layer = nn.ModuleList([Module1() for _ in range(2)])

    def forward(self, x):
        for layer in self.layer:
            x = layer(x)
        return x


module2 = Module2()

x = torch.rand(4,3)

y = module2(x)

print(x)
print(y)
```
