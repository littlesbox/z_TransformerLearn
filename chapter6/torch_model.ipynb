```python
import torch
from torch import nn

class simpleMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.m = nn.Sequential(
            nn.Linear(5,3),
            nn.ReLU(),
            nn.Linear(3,2),
        )

    def forward(self, x):
        return self.m(x)

model = simpleMLP()

x = torch.rand(3,5)

y = model(x)

torch.save(model.state_dict(), 'model_weights.pth')
torch.save(model, 'model.pth')
```

```python
x = torch.ones(3,5)
model = simpleMLP()
model.load_state_dict(torch.load('model_weights.pth'))
model.eval()
y = model(x)
print(y)
```
