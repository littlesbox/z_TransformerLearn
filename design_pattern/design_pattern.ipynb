```python
# 单例模式
class ConfigManager:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
    
    def __init__(self):
        self.name = "1"
        self.age = 0

    def set_name(self, name):
        if isinstance(name, str):
            self.name = name
        else:
            raise TypeError
        
    def set_age(self, age):
        if isinstance(age, int):
            self.age = age
        else:
            raise TypeError
        
    def set(self, item, value):
        if item == 'name':
            self.set_name(value)
        elif item == 'age':
            self.set_age(value)
        else:
            raise ValueError
        
    def get(self, item):
        if item == 'name':
            return self.name  
        elif item == 'age':
            return self.age
        else:
            raise ValueError

a = ConfigManager()
a.set('name', 'wang')
print(a.get('name'))
b = ConfigManager()
a is b
print(b.get('name'))
print(ConfigManager._instance)
```

```python
# 工厂模式
class BaseModel:
    def __init__(self):
        self.modelname = ""

    def train(self):
        pass

    def predit(self):
        pass

class BertModel(BaseModel):
    def __init__(self):
        self.modelname = 'bert'

    def train(self):
        print('bert is training')
    
    def predit(self):
        print('bert is preditting')

class GPTmodel(BaseModel):
    def __init__(self):
        self.modelname = 'gpt'

    def train(self):
        print('gpt is training')

    def predit(self):
        print('gpt is preditting')

class ModelFactory:
    model_dict = {'bert':BertModel, 'gpt':GPTmodel}

    @classmethod
    def create(cls, name):
        if name in cls.model_dict.keys():
            return cls.model_dict[name]()
        else:
            raise ValueError('model not found')

model = ModelFactory.create('bert')
model.train()
model.predit()
model = ModelFactory.create('gpt')
model.train()
model.predit()
```

```python
# 策略模式
# 不同的支付系统
class PaymentStrategy:

    @classmethod
    def pay(cls, amount):
        raise NotImplementedError
    

class Alipay(PaymentStrategy):

    @classmethod
    def pay(cls, amount):
        print(f'use Alipay cost {amount} yuan')


class WeiChatpay(PaymentStrategy):

    @classmethod
    def pay(cls, amount):
        print(f'use WeiChatpay cost {amount} yuan')


class Bankpay(PaymentStrategy):

    @classmethod
    def pay(cls, amount):
        print(f'use Bankpay cost {amount} yuan')


class PaymentContext:
    def __init__(self, strategy: PaymentStrategy):
        self._strategy = strategy

    def set_strategy(self, strategy:PaymentStrategy):
        self._strategy = strategy

    def pay(self, amount):
        if amount <= 0:
            raise ValueError("支付金额必须大于 0")
        self._strategy.pay(amount)

pay = PaymentContext(Bankpay)
pay.pay(100)
```

```python
# 策略模式
# 不同的导航策略
class BaseRoute:
    def __init__(self):
        pass

    def build_route(self, a, b):
        raise NotImplementedError
    

class RoadRoute(BaseRoute):
    def __init__(self):
        super().__init__()

    def build_route(self, a, b):
        print(f'使用公路导航从{a}到{b}')


class WalkingRoute(BaseRoute):
    def __init__(self):
        super().__init__()

    def build_route(self, a, b):
        print(f'使用步行导航从{a}到{b}')


class RouteContext:
    def __init__(self, strategy: BaseRoute):
        self._strategy = strategy
    
    def build_route(self, a, b):
        self._strategy.build_route(a,b)

route = RouteContext(RoadRoute())
route.build_route(1,2)
route = RouteContext(WalkingRoute())
route.build_route(1,2)
```

```python
# 观察者模式
# 新闻订阅系统
class Subscriber:

    def update(self, news: str):
        raise NotImplementedError
    

class UserSubscriber(Subscriber):
    def __init__(self, name):
        self.username = name

    def update(self, news: str):
        print(f'用户{self.username}订阅收到了新消息{news}')


class EmailSubscriber(Subscriber):
    def __init__(self, email):
        self.email = email

    def update(self, news: str):
        print(f'邮件{self.email}订阅收到了新消息{news}')


class NewsPublisher:
    def __init__(self):
        self.user = []


    def subscribe(self, user):
        if user not in self.user:
            self.user.append(user)
            print('已新增用户')
        else:
            print(f"订阅者已存在，跳过")


    def unsubscribe(self, user):
        if user in self.user:
            self.user.remove(user)
            print('已删除')
        else:
            print(f"未找到订阅者，无法移除")


    def notify(self, news: str):
        if not self.user:
            print("当前无订阅者，新闻未发送。")
            return

        for user in self.user:
            # 多态调用：具体行为由 subscriber 的实际类型决定
            user.update(news)


    def publish(self, news: str):
        print(f"\n发布新闻: {news}")
        self.notify(news)


# 创建发布者
publisher = NewsPublisher()

# 创建不同类型的订阅者
alice = UserSubscriber("Alice")
bob = UserSubscriber("Bob")
email_sub = EmailSubscriber("admin@example.com")

# 用户订阅新闻
publisher.subscribe(alice)
publisher.subscribe(bob)
publisher.subscribe(email_sub)


# 发布新闻 —— 所有订阅者自动收到通知
publisher.publish("Python 4.0 released")

# 某用户取消订阅
publisher.unsubscribe(bob)

# 再次发布 —— Bob 不再收到消息
publisher.publish("Django 5.0 announced")

# 尝试重复订阅（应有提示）
publisher.subscribe(alice)

    
```

```python
"""
装饰器模式 —— 咖啡系统示例
============================
演示如何通过“包装（wrap）”的方式在运行时动态为对象添加功能，
从而避免因组合而产生类爆炸。装饰器与被装饰对象实现同一个接口，
使得装饰器可以互相嵌套。
"""

# ============================================================
# 1. 抽象 Coffee 基类：定义核心接口
# ============================================================
class Coffee:
    """
    咖啡抽象基类。
    定义了所有咖啡（包括装饰器）必须提供的两个接口：
    - description() : 返回咖啡的描述
    - price()       : 返回咖啡的价格
    装饰器之所以也继承此类，是为了保证装饰后的对象依然是 Coffee，
    从而可以继续被其他装饰器包装。
    """

    def description(self) -> str:
        raise NotImplementedError

    def price(self) -> float:
        raise NotImplementedError
    
# ============================================================
# 2. 基础咖啡类：被装饰的核心对象
# ============================================================

class Americano(Coffee):
    """美式咖啡"""

    def description(self) -> str:
        return "Americano"

    def price(self) -> float:
        return 15.0


class Espresso(Coffee):
    """浓缩咖啡"""

    def description(self) -> str:
        return "Espresso"

    def price(self) -> float:
        return 12.0


# ============================================================
# 3. 装饰器抽象基类
# ============================================================
class CoffeeDecorator(Coffee):
    """
    装饰器基类。
    持有一个 Coffee 对象的引用，并在调用 description() 和 price() 时，
    将请求委托给该对象，同时可以加入自己的额外逻辑。
    注意：此类也继承自 Coffee，因此它本身也是一个“咖啡”，
    这是装饰器能够嵌套的关键。
    """

    def __init__(self, coffee: Coffee):
        self._coffee = coffee  # 被装饰的咖啡对象

    def description(self) -> str:
        # 默认委托给被装饰对象，具体装饰器可以在此基础上添加描述
        return self._coffee.description()

    def price(self) -> float:
        # 默认委托给被装饰对象，具体装饰器可以在此基础上累加价格
        return self._coffee.price()


# ============================================================
# 4. 具体装饰器：牛奶、糖、巧克力
# ============================================================

class MilkDecorator(CoffeeDecorator):
    """牛奶装饰器"""

    def description(self) -> str:
        # 在被装饰对象描述的基础上追加 " + Milk"
        return self._coffee.description() + " + Milk"

    def price(self) -> float:
        # 在被装饰对象价格的基础上增加牛奶的价格
        return self._coffee.price() + 5.0


class SugarDecorator(CoffeeDecorator):
    """糖装饰器"""

    def description(self) -> str:
        return self._coffee.description() + " + Sugar"

    def price(self) -> float:
        return self._coffee.price() + 2.0


class ChocolateDecorator(CoffeeDecorator):
    """巧克力装饰器"""

    def description(self) -> str:
        return self._coffee.description() + " + Chocolate"

    def price(self) -> float:
        return self._coffee.price() + 8.0



# 创建基础咖啡：美式
my_coffee = Americano()
print("基础咖啡:", my_coffee.description(), "| 价格:", my_coffee.price())

# 动态加牛奶
my_coffee = MilkDecorator(my_coffee)
print("加牛奶后:", my_coffee.description(), "| 价格:", my_coffee.price())

# 再加糖（可以继续嵌套）
my_coffee = SugarDecorator(my_coffee)
print("加糖后:", my_coffee.description(), "| 价格:", my_coffee.price())

# 再加巧克力
my_coffee = ChocolateDecorator(my_coffee)
print("加巧克力后:", my_coffee.description(), "| 价格:", my_coffee.price())

print("\n=== 一次性嵌套创建 ===")
# 也可以使用嵌套的方式直接创建
fancy_coffee = ChocolateDecorator(
    SugarDecorator(
        MilkDecorator(
            Espresso()
        )
    )
)
print(fancy_coffee.description(), "| 价格:", fancy_coffee.price())
```

```python
"""
适配器模式 —— 翻译系统示例
============================
演示如何通过适配器将两个不兼容的第三方翻译接口
转换为统一的 Translator 接口，从而让系统只依赖抽象，
不依赖具体第三方实现。
"""


# ============================================================
# 1. 模拟的第三方翻译库（接口不一致，不可修改）
# ============================================================

class BaiduTranslate:
    """
    模拟百度翻译 API（第三方库）。
    其方法名为 baidu_translate，参数为 content。
    我们无法修改这个类的代码。
    """
    def baidu_translate(self, content: str) -> str:
        # 模拟翻译处理，实际会调用百度翻译 API
        print(f"[百度翻译] 正在翻译: {content}")
        return f"百度翻译结果: {content}"  # 示意


class GoogleTranslate:
    """
    模拟谷歌翻译 API（第三方库）。
    其方法名为 google_trans，参数为 message。
    与 BaiduTranslate 的接口完全不同。
    """
    def google_trans(self, message: str) -> str:
        print(f"[谷歌翻译] 正在翻译: {message}")
        return f"Google translation: {message}"


# ============================================================
# 2. 系统内部期望的统一翻译接口
# ============================================================

class Translator:
    """
    翻译器抽象接口。
    系统内部的所有翻译操作都只依赖于这个接口。
    它定义了统一的 translate(text) 方法，
    任何适配器或实现都必须遵守。
    """
    def translate(self, text: str) -> str:
        raise NotImplementedError("子类必须实现 translate 方法")


# ============================================================
# 3. 适配器：将第三方接口适配为 Translator 接口
# ============================================================

class BaiduAdapter(Translator):
    """
    百度翻译适配器。
    内部持有 BaiduTranslate 对象，
    在 translate() 方法中调用 baidu_translate()，
    从而将不同的接口名称和参数适配为统一接口。
    """
    def __init__(self):
        # 持有第三方对象（可以通过依赖注入提高灵活性）
        self._baidu = BaiduTranslate()

    def translate(self, text: str) -> str:
        # 接口转换：translate -> baidu_translate
        return self._baidu.baidu_translate(text)


class GoogleAdapter(Translator):
    """
    谷歌翻译适配器。
    内部持有 GoogleTranslate 对象，
    在 translate() 方法中调用 google_trans()。
    """
    def __init__(self):
        self._google = GoogleTranslate()

    def translate(self, text: str) -> str:
        # 接口转换：translate -> google_trans
        return self._google.google_trans(text)


# ============================================================
# 4. 客户端代码：依赖统一接口
# ============================================================
def perform_translation(translator: Translator, text: str):
    """
    执行翻译操作的客户端函数。
    它完全依赖于 Translator 抽象，不知道具体的翻译服务商。
    这种设计使得切换翻译服务无需修改此函数。
    """
    result = translator.translate(text)
    print(f"翻译结果: {result}\n")


if __name__ == "__main__":
    # 需要百度翻译时，使用百度适配器
    baidu_translator = BaiduAdapter()
    perform_translation(baidu_translator, "你好世界")

    # 需要谷歌翻译时，使用谷歌适配器
    google_translator = GoogleAdapter()
    perform_translation(google_translator, "Hello world")

    # ----------------------------------------------------------
    # 关键思考解答（以注释形式体现）
    # ----------------------------------------------------------
    #
    # 思考1：什么东西“不兼容”？
    # 第三方翻译库提供了不同的方法名（baidu_translate vs google_trans）
    # 和不同的参数名（content vs message）。系统期望的是统一的 translate(text)，
    # 因此直接调用会导致客户端代码与第三方接口强耦合。
    #
    # 思考2：谁负责“转换接口”？
    # 适配器（Adapter）承担了这个职责。它包装了第三方对象，
    # 对外暴露 translate 方法，内部调用对应的第三方方法，
    # 从而完成接口的转换/对齐。
    #
    # 思考3：系统应该依赖谁？
    # 系统应该依赖抽象的 Translator 接口，而不是具体的第三方实现。
    # 通过依赖倒置，客户端代码（如 perform_translation）只与抽象交互，
    # 具体的翻译服务可以通过适配器灵活替换，符合开放封闭原则。
    #
    # 扩展性示例：新增微软翻译
    # class MicrosoftTranslate:  # 第三方
    #     def ms_translate(self, source_text):
    #         ...
    # class MicrosoftAdapter(Translator):
    #     def __init__(self):
    #         self._ms = MicrosoftTranslate()
    #     def translate(self, text):
    #         return self._ms.ms_translate(text)
    #
    # 使用：microsoft_translator = MicrosoftAdapter()
    # 无需修改 perform_translation 或任何已有代码。
    #
    # ----------------------------------------------------------
    # 适配器模式在实际开发中的应用场景：
    # - 将不同数据库驱动统一为 ORM 接口
    # - 将各种第三方支付 SDK 适配为统一的支付接口
    # - 包装遗留系统以适应新系统接口
    # - Python 的 with 语句上下文管理器实际也是一种适配器
    # ----------------------------------------------------------
```

```python
"""
代理模式 —— 图片加载系统示例
===============================
演示如何通过代理对象控制对真实资源的访问，实现延迟加载（Lazy Loading）。
客户端完全不知道面对的是代理还是真实对象，因为它们实现了相同的接口。
"""


# ============================================================
# 1. 抽象 Image 接口
# ============================================================
class Image:
    """
    图片抽象接口。
    定义了所有图片（真实图片和代理）必须提供的 display() 方法。
    客户端只依赖此接口，实现了“面向抽象编程”。
    """
    def display(self):
        raise NotImplementedError("子类必须实现 display 方法")


# ============================================================
# 2. 真实图片类 —— 创建成本高昂
# ============================================================
class RealImage(Image):
    """
    真实图片，代表存储在磁盘上的大图片文件。
    构造时即模拟从磁盘加载的耗时操作。
    """
    def __init__(self, filename: str):
        self._filename = filename
        self._load_from_disk()  # 立即加载（模拟耗时）

    def _load_from_disk(self):
        """模拟从磁盘加载图片的耗时操作（打印信息表示）。"""
        print(f"Loading image from disk... [{self._filename}]")

    def display(self):
        """显示图片（加载完成后才能调用）。"""
        print(f"Displaying {self._filename}")


# ============================================================
# 3. 代理类 —— 控制对真实对象的访问，实现延迟加载
# ============================================================
class ImageProxy(Image):
    """
    图片代理，作为 RealImage 的“占位符”。
    构造时不会创建 RealImage 实例（延迟加载），
    仅在第一次调用 display() 时才创建真实对象。
    后续调用直接委托给已创建的真实对象。
    """
    def __init__(self, filename: str):
        self._filename = filename
        self._real_image = None  # 开始时真实对象不存在

    def display(self):
        # 如果真实对象尚未创建，则现在创建（延迟加载）
        if self._real_image is None:
            print("首次调用 display，触发真实图片加载...")
            self._real_image = RealImage(self._filename)
        else:
            # 真实对象已存在，直接使用
            print("使用已加载的真实图片...")

        # 委托给真实对象执行显示操作
        self._real_image.display()


# ============================================================
# 4. 客户端使用示例
# ============================================================
if __name__ == "__main__":
    # 创建代理对象 —— 此时不会加载真实图片
    img_proxy = ImageProxy("cat.png")
    print("代理对象创建完毕，但真实图片尚未加载。\n")

    # 第一次调用 display() —— 触发加载并显示
    img_proxy.display()
    print("---")

    # 第二次调用 display() —— 直接使用已加载的对象，不再重复加载
    img_proxy.display()

    # ----------------------------------------------------------
    # 关键思考解答（注释形式）
    # ----------------------------------------------------------
    #
    # 思考1：为什么不直接创建 RealImage？
    # 因为真实对象的创建成本很高（模拟的磁盘加载），
    # 而有时客户端可能并不需要立即显示图片，或者根本不会使用。
    # 代理模式通过延迟加载来优化性能，避免不必要的资源消耗。
    #
    # 思考2：Proxy 和 RealImage 为什么都实现 Image 接口？
    # 因为这样客户端就可以无缝地使用代理替代真实对象，
    # 完全不需要知道自己面对的是代理还是真实对象。
    # 这种“透明性”是代理模式的核心特征。
    #
    # 思考3：谁负责控制真实对象创建？
    # ImageProxy 负责控制。它在内部持有 RealImage 引用，
    # 并在适当的时机（如第一次需要时）创建真实对象。
    # 除了延迟加载，代理还可以控制访问权限、记录日志、缓存结果等。
    #
    # 代理模式与装饰器模式的区别：
    # - 代理模式关注“控制对象访问”（如延迟加载、权限检查）
    # - 装饰器模式关注“动态增强对象功能”（如添加新行为）
    # 但两者在结构上非常相似，都实现了相同的接口并持有对象引用。
    #
    # 扩展应用场景：
    # - ORM 中查询集的延迟加载（Lazy Query）
    # - 网络代理（如 VPN、反向代理 nginx）
    # - 权限代理（检查用户权限后再调用真实服务）
    # - 缓存代理（先查缓存，不存在再请求远程）
    # - RPC / API Gateway（隐藏远程调用的细节）
    # ----------------------------------------------------------
```

```python
"""
外观模式 —— 电脑启动系统示例
=============================
演示如何通过外观（Facade）为复杂的子系统集合提供一个统一、
简化的接口。客户端无需了解内部 CPU、内存、磁盘的具体交互细节，
只需调用 computer.start() 即可完成启动流程。
"""


# ============================================================
# 1. 复杂的子系统：各自拥有独立且复杂的接口
# ============================================================

class CPU:
    """中央处理器子系统"""
    def freeze(self):
        print("CPU: 冻结处理器状态")

    def jump(self, position: str):
        print(f"CPU: 跳转到内存地址 {position}")

    def execute(self):
        print("CPU: 开始执行指令")


class Memory:
    """内存子系统"""
    def load(self, position: str, data: str):
        print(f"Memory: 将数据 '{data}' 加载到地址 {position}")


class Disk:
    """磁盘子系统"""
    def read(self, sector: str, size: int) -> str:
        print(f"Disk: 从扇区 {sector} 读取 {size} 字节数据")
        return "boot_sector_data"


# ============================================================
# 2. 外观类：封装复杂的子系统协作流程，提供统一入口
# ============================================================

class ComputerFacade:
    """
    电脑外观（Facade）。
    将启动电脑所需的复杂子流程封装在 start() 方法中，
    为客户端提供一个极简的“一键启动”接口。
    客户端完全不需要了解 CPU、内存、磁盘的调用顺序。
    """
    def __init__(self):
        # 外观内部组合了所有子系统（管理子系统的生命周期）
        self._cpu = CPU()
        self._memory = Memory()
        self._disk = Disk()

    def start(self):
        """统一的启动方法，封装了复杂的启动序列。"""
        print("ComputerFacade: 正在启动电脑...\n")
        
        # 步骤1：冻结 CPU 状态
        self._cpu.freeze()
        
        # 步骤2：从磁盘读取引导数据
        boot_data = self._disk.read(sector="0", size=512)
        
        # 步骤3：将引导数据加载到内存的特定位置
        self._memory.load(position="0x0000", data=boot_data)
        
        # 步骤4：CPU 跳转到内存引导位置
        self._cpu.jump(position="0x0000")
        
        # 步骤5：CPU 开始执行指令
        self._cpu.execute()
        
        print("\nComputerFacade: 电脑启动完毕。")


# ============================================================
# 3. 客户端代码：极简调用
# ============================================================
if __name__ == "__main__":
    # 客户端只需创建外观并调用 start()
    computer = ComputerFacade()
    computer.start()
    
    # 如果没有外观，客户端必须手动依次调用：
    # cpu.freeze()
    # boot_data = disk.read("0", 512)
    # memory.load("0x0000", boot_data)
    # cpu.jump("0x0000")
    # cpu.execute()
    # 这既繁琐又暴露了子系统的复杂性，且顺序错误可能导致启动失败。

    # ----------------------------------------------------------
    # 关键思考解答
    # ----------------------------------------------------------
    #
    # 思考1：Facade 为什么存在？
    # 因为子系统（CPU、Memory、Disk）的操作复杂且必须按正确顺序执行，
    # 客户端直接操作容易出错且增加认知负担。外观模式通过提供统一入口
    # start() 隐藏了复杂性，让客户端只需关心“启动”这个高层次操作。
    #
    # 思考2：Facade 是否替代子系统？
    # 不替代。子系统依然独立存在，并且可以被直接访问（如果需要）。
    # Facade 只是提供了一层包装，简化了最常见的使用场景。
    # 如果客户端需要更精细的控制，仍然可以绕过外观直接操作子系统。
    #
    # 思考3：Facade 和 Adapter 有什么区别？
    # - Adapter（适配器）的核心是“接口转换”，使得原本不兼容的接口
    #   能够一起工作。它通常用于整合旧的或第三方接口。
    # - Facade（外观）的核心是“接口简化”，它为一组复杂的接口提供
    #   一个高层、统一的接口。它不解决接口不兼容的问题，而是解决
    #   接口太复杂的问题。两者目的不同。
    #
    # 实际应用场景：
    # - HuggingFace pipeline：隐藏了分词、模型加载、推理的复杂流程
    # - Docker CLI：一条 docker run 命令背后是复杂的容器生命周期管理
    # - Git：git checkout 封装了底层对象操作
    # - ORM Session：隐藏了连接池、事务、SQL 生成的复杂性
    # - 电商下单接口：统一封装库存检查、支付、物流等微服务调用
    # ----------------------------------------------------------
```

```python

"""
模板方法模式 —— 数据处理框架示例
================================
演示如何通过父类固定算法骨架（处理流程），而将具体步骤的实现延迟到子类。
客户端只需调用统一的 process() 方法，流程顺序不可更改，
但每个步骤的具体行为可由不同子类定制。
"""


# ============================================================
# 1. 抽象父类：定义算法骨架（模板方法）
# ============================================================
class DataProcessor:
    """
    数据处理器抽象类。
    定义了数据处理的标准流程（模板方法 process()），
    该流程的顺序是固定的，不可被子类修改。
    每个步骤的具体实现由子类提供（通过抽象方法或钩子方法）。
    """

    def process(self):
        """
        模板方法：定义数据处理的标准流程。
        流程顺序固定为：读取 -> 清洗 -> 分析 -> 输出
        （子类不能修改此流程，只能实现具体步骤）
        """
        data = self.read_data()
        data = self.clean_data(data)
        result = self.analyze_data(data)
        self.export_result(result)

    # ---- 抽象步骤：必须由子类实现 ----
    def read_data(self):
        """读取数据"""
        raise NotImplementedError

    def clean_data(self, data):
        """清洗数据（可被子类重写，提供默认实现时数据不做任何改变）"""
        # 默认不做清洗，子类可根据需要重写
        return data

    def analyze_data(self, data):
        """分析数据"""
        raise NotImplementedError

    def export_result(self, result):
        """输出结果"""
        raise NotImplementedError


# ============================================================
# 2. 具体子类：CSV 数据处理器
# ============================================================
class CSVProcessor(DataProcessor):
    """处理 CSV 文件的具体实现"""

    def read_data(self):
        print("CSVProcessor: 读取 CSV 文件数据...")
        # 模拟返回的数据（实际代码会使用 csv 模块）
        return [{"name": "Alice", "score": 85},
                {"name": "Bob", "score": 92}]

    def clean_data(self, data):
        print("CSVProcessor: 清洗 CSV 数据（去除空行）...")
        # 模拟清洗操作
        return data

    def analyze_data(self, data):
        print("CSVProcessor: 分析 CSV 数据（计算平均分）...")
        total = sum(item["score"] for item in data)
        avg = total / len(data)
        return {"average_score": avg}

    def export_result(self, result):
        print(f"CSVProcessor: 输出结果 -> 平均分: {result['average_score']:.2f}")


# ============================================================
# 3. 具体子类：JSON 数据处理器
# ============================================================
class JSONProcessor(DataProcessor):
    """处理 JSON 文件的具体实现"""

    def read_data(self):
        print("JSONProcessor: 读取 JSON 文件数据...")
        # 模拟 JSON 数据
        import json
        return json.loads('[{"product": "apple", "price": 5}, {"product": "banana", "price": 3}]')

    # 注意：JSONProcessor 没有重写 clean_data，
    # 因此使用父类默认的空清洗实现，体现了模板方法模式的灵活性。

    def analyze_data(self, data):
        print("JSONProcessor: 分析 JSON 数据（统计商品数量）...")
        count = len(data)
        total_price = sum(item["price"] for item in data)
        return {"product_count": count, "total_price": total_price}

    def export_result(self, result):
        print(f"JSONProcessor: 输出结果 -> 商品数: {result['product_count']}, 总价: {result['total_price']}")


# ============================================================
# 4. 客户端使用示例
# ============================================================
if __name__ == "__main__":
    # 客户端使用 CSV 处理器
    csv_processor = CSVProcessor()
    print("=== 开始处理 CSV 数据 ===")
    csv_processor.process()
    # 流程固定，输出顺序严格：读取 -> 清洗 -> 分析 -> 输出

    print("\n=== 开始处理 JSON 数据 ===")
    # 客户端使用 JSON 处理器
    json_processor = JSONProcessor()
    json_processor.process()

    # ----------------------------------------------------------
    # 关键思考解答（以注释形式体现）
    # ----------------------------------------------------------
    #
    # 思考1：什么东西固定？
    # 数据处理的主流程（读取 -> 清洗 -> 分析 -> 输出）是固定的，
    # 这个顺序是业务逻辑的核心，不能随意更改。
    # 模板方法模式将这个固定骨架放在父类中，保证一致性。
    #
    # 思考2：什么东西变化？
    # 每个步骤的具体实现（如何读取 CSV 还是 JSON、如何分析等）
    # 会因数据类型不同而变化。模板方法将变化的细节推迟到子类实现。
    #
    # 思考3：为什么父类控制流程？
    # 父类控制流程体现了“控制反转”（IoC）的思想：
    # 流程的控制权在父类手中，子类只需填充具体步骤。
    # 这样避免了每个子类重复编写相同的流程控制代码，
    # 且算法骨架稳定，不易出错。
    #
    # 模板方法模式的实际应用场景：
    # - PyTorch/HuggingFace 的 Trainer：训练流程（前向、反向、优化）固定，
    #   但模型、损失函数、优化器等可变。
    # - Django 的类视图：请求处理流程固定（dispatch -> get/post 等），
    #   开发者只需重写 get/post 方法。
    # - sklearn 的 Pipeline：fit/predict 流程固定，但每步的变换器可变。
    # - 饮料制作：烧水 -> 冲泡 -> 倒入杯子 -> 加调料，具体饮品不同。
    #
    # ---------------------------------------------------------
```

```python
"""
责任链模式 —— 请假审批系统示例
================================
演示如何将一系列处理者（审批人）串联成链，请求在链上自动传递，
直到某个处理者处理为止。从而消除复杂的 if-else 分支，
实现请求发送者与处理者的解耦。
"""


# ============================================================
# 1. 请求类：封装需要处理的数据
# ============================================================
class LeaveRequest:
    """请假请求，包含员工姓名和请假天数"""
    def __init__(self, employee_name: str, days: int):
        self.employee_name = employee_name
        self.days = days

    def __str__(self):
        return f"{self.employee_name} 请假 {self.days} 天"


# ============================================================
# 2. 抽象处理者：定义接口和链的构建方法
# ============================================================
class Approver:
    """
    审批人抽象类。
    每个审批人有一个可选的下一级审批人，构成链。
    定义 handle_request 方法，子类实现具体审批逻辑。
    """
    def __init__(self, title: str):
        self._title = title          # 审批人头衔
        self._next_approver = None   # 下一个审批人（链的下一个节点）

    def set_next(self, approver: "Approver") -> "Approver":
        """
        设置链的下一个处理者，并返回下一个处理者，
        这样可以实现链式调用构建链：leader.set_next(manager).set_next(director)
        """
        self._next_approver = approver
        return approver

    def handle_request(self, request: LeaveRequest):
        """
        处理请求的默认实现：如果当前节点能处理则处理，
        否则传递给下一个节点。
        子类可以重写此方法或提供 can_handle 抽象方法。
        """
        if self.can_handle(request):
            self.process(request)
        elif self._next_approver is not None:
            print(f"{self._title} 无权限审批 {request.days} 天，转交上级...")
            self._next_approver.handle_request(request)
        else:
            # 链末尾，无人能处理
            print(f"无法审批：{request}，请设置更高级别审批人。")

    def can_handle(self, request: LeaveRequest) -> bool:
        """判断当前审批人是否有权处理该请求（子类实现）"""
        raise NotImplementedError

    def process(self, request: LeaveRequest):
        """具体处理逻辑（子类实现）"""
        raise NotImplementedError


# ============================================================
# 3. 具体处理者：组长、经理、总监
# ============================================================

class TeamLeader(Approver):
    """组长：有权审批 3 天及以下的请假"""
    def __init__(self):
        super().__init__("组长")

    def can_handle(self, request: LeaveRequest) -> bool:
        return request.days <= 3

    def process(self, request: LeaveRequest):
        print(f"[组长审批] 同意 {request}")


class Manager(Approver):
    """经理：有权审批 7 天及以下的请假"""
    def __init__(self):
        super().__init__("经理")

    def can_handle(self, request: LeaveRequest) -> bool:
        return request.days <= 7

    def process(self, request: LeaveRequest):
        print(f"[经理审批] 同意 {request}")


class Director(Approver):
    """总监：有权审批 7 天以上的请假（无上限）"""
    def __init__(self):
        super().__init__("总监")

    def can_handle(self, request: LeaveRequest) -> bool:
        return True  # 总监可以审批任何天数

    def process(self, request: LeaveRequest):
        print(f"[总监审批] 同意 {request}")


# ============================================================
# 4. 客户端使用示例
# ============================================================
if __name__ == "__main__":
    # 构建责任链：组长 -> 经理 -> 总监
    leader = TeamLeader()
    manager = Manager()
    director = Director()

    # 组装链
    leader.set_next(manager).set_next(director)

    # 也可以直接指定链的起点为 leader

    # 创建不同天数的请假请求并提交
    requests = [
        LeaveRequest("Alice", 2),   # 组长审批
        LeaveRequest("Bob", 5),     # 经理审批
        LeaveRequest("Charlie", 8), # 总监审批
        LeaveRequest("Diana", 3),   # 组长审批
    ]

    print("=== 请假审批流程 ===")
    for req in requests:
        print(f"\n提交请求：{req}")
        leader.handle_request(req)  # 从链的头部开始传递

    # ----------------------------------------------------------
    # 关键思考解答
    # ----------------------------------------------------------
    #
    # 思考1：什么东西在“传递”？
    # 请假请求（LeaveRequest）在审批人链上传递。
    # 每个审批人都可能处理它，如果不能处理就交给下一个。
    #
    # 思考2：为什么不能写大量 if-else？
    # 因为审批规则会变化（如新增 HR、CEO 等审批人），
    # 使用 if-else 会导致代码臃肿且难以维护，违反开闭原则。
    # 责任链模式将每个规则封装为独立的处理者节点，
    # 可以动态组合链的结构，增减节点不影响已有代码。
    #
    # 思考3：谁决定“是否继续传递”？
    # 每个处理者自己决定是否能处理请求（can_handle），
    # 如果不能处理，则显式传递给下一个处理者。
    # 这样责任链的流转逻辑分布在各个节点中。
    #
    # 实际应用场景：
    # - Django/FastAPI 的中间件链（请求依次经过各个中间件处理）
    # - Spring Security 的过滤器链
    # - 审批工作流（OA 系统）
    # - 敏感词过滤链
    # - 日志处理链（按级别传递给不同日志记录器）
    #
    # 责任链模式的优点：
    # - 降低耦合度：请求发送者无需知道哪个处理者会处理请求
    # - 增强灵活性：可以动态增加或重新排序链中的节点
    # - 符合单一职责原则：每个处理者只需关注自己的处理逻辑
    # ----------------------------------------------------------
```
