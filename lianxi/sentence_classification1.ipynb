```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TextClassificationPipeline, AutoModel
model1 = AutoModelForSequenceClassification.from_pretrained('uer/roberta-base-finetuned-dianping-chinese')
# print(model1)
print("\n")
model2 = AutoModel.from_pretrained('uer/roberta-base-finetuned-dianping-chinese')
# print(model2)

tokenizer = AutoTokenizer.from_pretrained('uer/roberta-base-finetuned-dianping-chinese')
seq = '今天天气真好啊'
inputs = tokenizer(seq, return_tensors='pt')
# print(inputs)

outputs1 = model1(**inputs)
outputs2 = model2(**inputs)
print(outputs1)
print("\n")
print(outputs2['last_hidden_state'].shape)
```

```python
import torch.nn as nn

class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.bert = model2
        self.m = nn.Linear(768,5)

    def forward(self,x):
        x = self.bert(**x)
        x = self.m(x.last_hidden_state[:,0,:])
        
        return x

model3 = MyModel()

outputs3 = model3(inputs)
print(outputs3)
```
